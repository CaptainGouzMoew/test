```python
def check_null_case(name, address):
    # 5 Cases:
    # null name & address: 1
    # null name only: 2
    # null address only: 3
    # fulll name & address: 4
    # other: 0 (this should never happend)
    
    name = name.strip()
    address = address.strip()
    if len(name) == 0 and len(address) == 0:
        return 1
    elif len(name) == 0:
        return 2
    elif len(address) == 0:
        return 3
    elif len(name) > 0 and len(address) > 0:
        return 4
    else:
        return 0
    
hms_df_all_cases_proba['null_case'] = hms_df_all_cases_proba.fillna('').apply(lambda x: check_null_case(x['customer_name'], x['customer_address']), axis=1)
hms_df_all_cases_proba
```

```python
import unicodedata
import re

def remove_accents(input_str):
    if not isinstance(input_str, str):
        return input_str
    input_str = input_str.replace('Đ', 'D').replace('đ', 'd')
    normalized_str = unicodedata.normalize('NFD', input_str)
    stripped_str = ''.join(
        char for char in normalized_str if unicodedata.category(char) != 'Mn'
    )
    return stripped_str

def get_letter_cluster(x):
    x = remove_accents(x)
    if x:
        if ' ' in x:
            first_character = x.strip().split(' ')[1][0].lower()
        else:
            first_character = x.strip()[0].lower()  
        if re.match(r'[a-z]', first_character):  
            return first_character
    return 26



def get_name_cluster(x, eng_lst = ['w', 'y', 'z', 'j', 'f'], vn_lst = ['t', 'v', 'h', 'n', 'd', 'l', 'p']):
    x = x.strip().lower()
    # Get first value of string
    n0 = remove_accents(x[0])
    
    # If first value isnt a vietnamese letter then return blank
    if n0 in eng_lst or not re.match(r'[a-v]', n0):
        return ''
    
    # If first letter is NOT a common letter in vietnamese alphabet then return it
    if n0 not in vn_lst:
        return n0
    
    # If first letter is a common letter in vietnamese alphabet then process it 
    # EXAMPLE:
    # NGUYEN VAN AN => NA
    # NGUYEN => N
    # NGUYEN あn => N
    # Check if x is a single word
    if ' ' not in x:
        return n0

    # If it's multi word
    x_s = x.split(' ')
    n1 = remove_accents(x_s[-1][0])

    # Check if second letter is an english letter or not
    if n1 in eng_lst or not re.match(r'[a-z]', n1):
        return n0
    else:
        return n0 + n1

def cluster_to_dict(cluster_lst):
    dct = {}
    for i in range(len(cluster_lst)):
        dct[cluster_lst[i]] = i
    return dct
    
case3 = hms_df_all_cases_proba[hms_df_all_cases_proba['null_case'] == 3]
case3['cluster'] = case3['customer_name'].apply(lambda x: get_name_cluster(x))
dct = cluster_to_dict(case3['cluster'].drop_duplicates().sort_values().tolist())
case3['cluster'] = case3['cluster'].apply(lambda x: dct[x] )

case3
```
